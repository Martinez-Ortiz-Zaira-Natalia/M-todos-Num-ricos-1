# Gr치fica de $f(x)=x - 3^{-x}$

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

def g(x):
    return x
def h(x):
    return -3**(-x)
# Funci칩n cuadr치tica.
def f(x):
    return x - 3**(-x)
```

```python
# Valores del eje X que toma el gr치fico.
x=np.linspace(-10,10,100)
```

```python
# Graficar ambas funciones.
plt.figure(figsize=(5 , 5))
plt.plot(x, f(x))
plt.plot(x, g(x))
plt.plot(x, h(x))
plt.axhline(0, color="black")
plt.axvline(0, color="black")
plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.savefig("output.png") # Guardar gr치fico como im치gen PNG.
# Mostrarlo.


plt.grid(color="gray")
plt.title("Grafica de $f(x)=x - 3**(-x)$")
plt.xlabel("Eje X")
plt.ylabel("Eje Y")
```

En la Gr치fica nos explica como la funci칩n f(x)
es una suma de funciones las cuales son g(x) + h(x)
