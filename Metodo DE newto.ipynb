```python
#Definimos las bibliotecas a usar
import sympy as sp
import numpy as np
```

```python
# Variables
x1, x2 = sp.symbols('x1 x2')

#  Funcion a minimizar(aqui se puede cambiar)
f = 100*(x2 - x1**2)**2 + (1 - x1)**2
# Punto inicial
x0 = [-1.2, 1.0]
```

```python
# Gradiente
grad = sp.Matrix([
    sp.diff(f, x1),
    sp.diff(f, x2)
])
print("Gradiente:")
sp.pprint(grad)
```

```python
# Hessiano
H = grad.jacobian([x1, x2])

print("\nHessiano:")
sp.pprint(H)
```

```python
# Funciones
grad_fun = sp.lambdify((x1, x2), grad, "numpy")
H_fun = sp.lambdify((x1, x2), H, "numpy")
f_fun = sp.lambdify((x1, x2), f, "numpy")
```

```python
def newton(x0, tol=1e-8, max_iter=2):

    x = np.array(x0, dtype=float)

    print("\nIter |      x1      |      x2      |     f(x)")
    print("-"*55)

    for k in range(max_iter):

        g = np.array(grad_fun(x[0], x[1]), dtype=float).flatten()
        Hn = np.array(H_fun(x[0], x[1]), dtype=float)

        p = np.linalg.solve(Hn, -g)

        x = x + p

        print(f"{k+1:4d} | {x[0]:11.6f} | {x[1]:11.6f} | {f_fun(x[0],x[1]):12.6f}")

        if np.linalg.norm(g) < tol:
            break

    return x
```

```python
sol = newton(x0, max_iter=2)

print("\nSolución encontrada:")
print(sol)
```
